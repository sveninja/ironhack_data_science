```python
import pandas as pd
import requests
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
url = 'https://api.coingecko.com/api/v3/coins/bitcoin/market_chart/range?vs_currency=usd&from=1558224000&to=1560816000'
```

```python
result = requests.get(url)
result
```

```python
data = result.json()
data
```

```python
df_data = pd.DataFrame(data)
df_data.head(10)
```

```python
len(df_data)
```

```python
# Bitcoin, vs. USD, 1 Month
# request with explicit paramenters dictionary

# url = base + 'coins/bitcoin/market_chart'
param_dict = { 'id':'bitcoin', 'vs_currency':'usd', 'days':30}

result = requests.get(url, params=param_dict)

j = result.json()
df = pd.DataFrame(j['prices'], columns=['time', 'price'])
df.head(10)
```

```python
df = pd.DataFrame(j['prices'], columns=['time', 'price'])
df['time'] = pd.to_datetime(df['time'], unit='ms')

df.set_index('time', inplace=True)
df.head(10)
```

```python
df.plot()
```
