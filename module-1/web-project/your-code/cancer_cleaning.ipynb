```python
df.year_of_birth = df.year_of_birth.fillna(0)
df.year_of_death = df.year_of_death.fillna(0)
```

```python
df.year_of_birth = df.year_of_birth.astype(int)
df.year_of_death = df.year_of_death.astype(int)
```

```python
for col in df:
    print(col)
    print(df[col].unique())
    print()
```

```python
# assigning "missing" to missing data in gender column (nan, unknown, not reported)
lst_gender_new = []
for entry in df.gender:
    if entry not in ['male', 'female']:
        lst_gender_new.append('missing')
    else:
        lst_gender_new.append(entry)
df.gender = lst_gender_new

# assigning "missing" to missing data in race column
lst_race = ['white', 'asian', 'black or african american', 'american indian or alaska native', 'native hawaiian or other pacific islander', 'other']
lst_race_new = []
for entry in df.race:
    if entry not in lst_race:
        lst_race_new.append('missing')
    else:
        lst_race_new.append(entry)
df.race = lst_race_new
        
# assigning "missing" to missing data in ethnicity column
lst_ethnicity = ['not hispanic or latino', 'hispanic or latino']
lst_ethnicity_new = []
for entry in df.ethnicity:
    if entry not in lst_ethnicity:
        lst_ethnicity_new.append('missing')
    else:
        lst_ethnicity_new.append(entry)
df.ethnicity = lst_ethnicity_new

# assigning "missing" to missing data in state column
lst_state = ['released']
lst_state_new = []
for entry in df.state:
    if entry not in lst_state:
        lst_state_new.append('missing')
    else:
        lst_state_new.append(entry)
df.state = lst_state_new

# assigning "missing" to missing data in height column
#df['height'] = df['height'].replace(np.nan, 'missing', regex=True)


```

```python
df.height = df.height.round(1)
print(df.height.unique())
df.weight = df.weight.round(1)
print(df.weight.unique())
df.bmi = df.bmi.round(1)
print(df.bmi.unique())
```
